```python
import csv
import xlsxwriter
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get('https://aqarmap.com.eg/en/for-sale/property-type/?default=1')
response.text
```

```python
soup =BeautifulSoup(response.text,'html.parser')#turn the page into HTML form
```

```python
page=soup.find('div',class_="search-listing-card")
page
```

```python
price=soup.find('div',class_='search-listing-card__price').text.split("\n")
price[1]
price=price[1]
price
```

```python
location=soup.find('p',class_="search-listing-card__address").find('span').text
location
```

```python
area=soup.find('label',class_="search-listing-card__attributes_item").text.replace(' ','').replace('\n','')
area
```

```python
for i in range(1, 5):#loop to loop into pages(5)
    response=requests.get('https://aqarmap.com.eg/en/for-sale/property-type/?default=1&page=' + str(i))
    soup =BeautifulSoup(response.text,'html.parser')
   
    for page in soup.find_all('div',class_="search-listing-card"):#to loop in one pag
        price=page.find('div',class_='search-listing-card__price').text.split("\n")
        price=price[1]+price[2]
        location=page.find('p',class_="search-listing-card__address").find('span').text
        area=page.find('label',class_="search-listing-card__attributes_item").text.replace(' ','').replace('\n','')
        print(f'Price: {price}, Location: {location}, Area: {area}')
```

```python
header=["Location","price","Area"]
x=open("aqarmap.csv",'w',encoding="utf-8-sig")
writer=csv.writer(x)
writer.writerow(header)
for i in range(1, 5):#this loop for looping in pages 
    response=requests.get('https://aqarmap.com.eg/en/for-sale/property-type/?default=1&page=' + str(i))
    soup =BeautifulSoup(response.text,'html.parser')
    for page in soup.find_all('div',class_="search-listing-card"):#this loop for looping in one page
        price=page.find('div',class_='search-listing-card__price').text.split("\n")
        price=price[1]+price[2]
        location=page.find('p',class_="search-listing-card__address").find('span').text
        area=page.find('label',class_="search-listing-card__attributes_item").text.replace(' ','').replace('\n','')
        writer.writerow([location,price,area])
x.close()
```
